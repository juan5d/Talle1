#### 1. Si x0 = 5 y xn = 3xn−1 mod 150, determine x1,...,x10####

```python
def mixedMethod(x, a, mod):
 
    periodo = 1
    bandera = 0
    val=0.0 
    while(periodo <=10):
        
        
        if (periodo == 0):
            bandera = x
        x = (a * x) % mod
        
        val=(x*1.0)/(mod*1.0)
        if(periodo == mod):
            print("El periodo es completo: ", periodo)
        else:
            print("El periodo es incompleto:", periodo)
 
        periodo = periodo + 1
        print(val)
    
def main():
    x = 5
    a = 3
    m = 150
    mixedMethod(x,a,m)
 
if __name__ == "__main__":
    main()
```
